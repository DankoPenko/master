```python
import random
from pathlib import Path 
import os
import json
directory = 'C:\Stuff\Dataset-creation\data\stump-original-offset-5-random'
f = open(os.path.join(directory, 'transforms.json'))
file = json.load(f)
new_frames = []

train_frames = [frame for frame in file['frames'] if 'train' in frame['file_path']]
one_sixth_count_train = len(train_frames) // 6
random_selected_train_frames = random.sample(train_frames, one_sixth_count_train)

for i, frame in enumerate(file['frames']):
    old_name = frame['file_path']
    parts = old_name.split("/")
    if 'eval' in frame['file_path']:
        new_frames.append(frame)
    else:
        if frame in random_selected_train_frames:
            new_frames.append(frame)


file['frames'] = new_frames
transforms_path = os.path.join(directory, 'transforms.json')
with open(transforms_path, "w") as jsonFile:
    json.dump(file, jsonFile, indent=4)
```
