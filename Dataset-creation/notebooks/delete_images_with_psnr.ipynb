```python
import os

def delete_images_with_psnr_in_subfolders(main_folder):
    for root, dirs, files in os.walk(main_folder):
        for filename in files:
            file_path = os.path.join(root, filename)
            if 'psnr' in filename:
                try:
                    os.remove(file_path)
                    print(f"Deleted: {file_path}")
                except Exception as e:
                    print(f"Error deleting {file_path}: {e}")

main_folder = 'C:\Stuff\Dataset-creation\\renders'
delete_images_with_psnr_in_subfolders(main_folder)
```
