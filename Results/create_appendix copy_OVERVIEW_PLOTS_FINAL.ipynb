```python
def extract_number(filename):
    match = re.search(r'(\d+\.?\d*)-000000', filename)
    return float(match.group(1)) if match else 0  # Convert the extracted number to float
```

```python
import os
import re
import numpy as np
path_txts = r'C:\Stuff\Results\appendix_texts'
path_plots = r'C:\Stuff\Results\gigaplots'

latex_code = ""
for folder in os.listdir(path_plots):
    latex_code = ""
    path = os.path.join(path_plots, folder)
    for subfolder in os.listdir(path):
        subfolder_path = os.path.join(path, subfolder)
        if os.path.isdir(subfolder_path):
            # Start a new figure environment for each subfolder

            latex_code += "\\begin{figure}[h]\n    \\centering\n"
            
            # Predefined subfigure for the "original" image at the start of each subfolder
            latex_code += """\\begin{subfigure}{0.16\\textwidth}
        \\includegraphics[width=\\linewidth]{images/appendix_plots/nerfacto_stump_opt_off/nerfacto_stump_opt_off_benchmark-000000-img.jpg} 
        \\caption{orig.}
        \\label{fig:nerfacto_stump_opt_off_benchmark-000000-img.jpg}
        \\end{subfigure}\\hfil % <-- added
        """
            
            filenames = os.listdir(subfolder_path)
            sorted_files = sorted(filenames, key=extract_number)
            # print(sorted_files)
            selected_filenames = np.linspace(0, len(sorted_files) - 1, 5, dtype=int)  # Select 5 evenly spaced indices
            
            for i in selected_filenames:
                file1 = sorted_files[i]
                caption = 321  # Default caption (You might want to modify this to be more descriptive)
                match = re.search(r"-(?!000000)(\d+\.?\d*)-", file1)
                if match is not None:
                    caption = float(match.group(1))
                if caption >= 1.0:
                    caption = int(caption)
                
                folder = folder.replace(' ', '')
                subfigure_code = f"""\\begin{{subfigure}}{{0.16\\textwidth}}
        \\includegraphics[width=\\linewidth]{{images/appendix_plots/{folder}/{file1}}} 
        \\caption{{{caption}}}
        \\label{{fig:{file1}}}
        \\end{{subfigure}}\\hfil % <-- added
        """
                latex_code += subfigure_code
            
            latex_code += "\\medskip\n"  # Adding a new line with \medskip at the end of each subfolder's content
            # Adding a figure caption that describes the subfolder content
            figure_caption = f"\\caption{{Images from the {subfolder} folder.}}\n"
            latex_code += figure_caption
            latex_code += "\\end{figure}\n"  # Close the figure environment for the folder

    latex_code += "%====================================% \n"
    with open(f'{os.path.join(path_txts, folder)}.txt', 'w') as file: file.write(latex_code)
    # print(folder, subfigure_code)
```

```python
import os
import shutil
path_plots = r'C:\Stuff\Results\gigaplots'
path_dump_plots = r'C:\Stuff\Results\gigaplots_dump'
for folder in os.listdir(path_plots):
    source_folder = os.path.join(path_plots, folder)
    destination_folder = os.path.join(path_dump_plots, folder)
    os.makedirs(destination_folder, exist_ok=True)

    for subfolder in os.listdir(source_folder):
        filenames = os.listdir(os.path.join(source_folder, subfolder))
        sorted_files = sorted(filenames, key=extract_number)
        selected_filenames = np.linspace(0, len(sorted_files) - 1, 5, dtype=int)  # Select 5 evenly spaced indices
        for index in selected_filenames:
            if sorted_files[index].endswith(".jpg"):
                source_file = os.path.join(source_folder, subfolder, sorted_files[index])
                destination_file = os.path.join(destination_folder, sorted_files[index])
                shutil.copyfile(source_file, destination_file)
        
```
