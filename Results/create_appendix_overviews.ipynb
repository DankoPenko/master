```python
def extract_number(filename):
    match = re.search(r'(\d+\.?\d*)-000000', filename)
    return float(match.group(1)) if match else 0  # Convert the extracted number to float
```

```python
import re
import os
import numpy as np

path = r'C:\Stuff\Results\gigaplots\ nerfacto_stump_opt_off'
latex_code = "\\begin{figure}[h]\n    \\centering\n"

for folder in os.listdir(path):
    folder_path = os.path.join(path, folder)
    if os.path.isdir(folder_path):
        filenames = os.listdir(folder_path)
        # if 'benchmark' not in filenames[0]:
        #   sorted_files = sorted(filenames, key=extract_number)
        # else:
        #   print(sorted_files)
        sorted_files = sorted(filenames, key=extract_number)
        selected_filenames = np.linspace(0, len(sorted_files) - 1, 5, dtype=int)  # Select 5 evenly spaced indices
        # Adding predefined subfigure for the original image at the start of each folder
        latex_code += """\\begin{subfigure}{0.16\\textwidth}
      \\includegraphics[width=\\linewidth]{images/appendix_plots/nerfacto_stump_opt_off/nerfacto_stump_opt_off_benchmark-000000-img.jpg} 
      \\caption{orig.}
      \\label{fig:nerfacto_stump_opt_off_benchmark-000000-img.jpg}
    \\end{subfigure}\\hfil % <-- added
    """
        
        for i in selected_filenames:
            file1 = sorted_files[i]
            caption = 321  # Default caption (You might want to modify this to be more descriptive)
            match = re.search(r"-(?!000000)(\d+\.?\d*)-", file1)
            if match is not None:
                caption = float(match.group(1))
            if caption >= 1.0:
                caption = int(caption)
            
            subfigure_code = f"""\\begin{{subfigure}}{{0.16\\textwidth}}
      \\includegraphics[width=\\linewidth]{{images/appendix_plots/nerfacto_stump_opt_off/{file1}}} 
      \\caption{{{caption}}}
      \\label{{fig:{file1}}}
    \\end{{subfigure}}\\hfil % <-- added
    """
            latex_code += subfigure_code
        latex_code += "\\medskip\n"  # Adding a new line with \medskip at the end of each folder

latex_code += "%====================================% \n\\end{figure}"
print(latex_code)
```

```python
import os
import re
import numpy as np

latex_code = ""

for folder in os.listdir(path):
    folder_path = os.path.join(path, folder)
    if os.path.isdir(folder_path):
        # Start a new figure environment for each folder
        latex_code += "\\begin{figure}[h]\n    \\centering\n"
        
        # Predefined subfigure for the "original" image at the start of each folder
        latex_code += """\\begin{subfigure}{0.16\\textwidth}
      \\includegraphics[width=\\linewidth]{images/appendix_plots/nerfacto_stump_opt_off/nerfacto_stump_opt_off_benchmark-000000-img.jpg} 
      \\caption{orig.}
      \\label{fig:nerfacto_stump_opt_off_benchmark-000000-img.jpg}
    \\end{subfigure}\\hfil % <-- added
    """
        
        filenames = os.listdir(folder_path)
        selected_filenames = np.linspace(0, len(filenames) - 1, 5, dtype=int)  # Select 5 evenly spaced indices
        sorted_files = sorted(filenames, key=extract_number)
        selected_filenames = np.linspace(0, len(sorted_files) - 1, 5, dtype=int)  # Select 5 evenly spaced indices

        for i in selected_filenames:
            file1 = sorted_files[i]
            caption = 321  # Default caption (You might want to modify this to be more descriptive)
            match = re.search(r"-(?!000000)(\d+\.?\d*)-", file1)
            if match is not None:
                caption = float(match.group(1))
            if caption >= 1.0:
                caption = int(caption)
            
            subfigure_code = f"""\\begin{{subfigure}}{{0.16\\textwidth}}
      \\includegraphics[width=\\linewidth]{{images/appendix_plots/nerfacto_stump_opt_off/{file1}}} 
      \\caption{{{caption}}}
      \\label{{fig:{file1}}}
    \\end{{subfigure}}\\hfil % <-- added
    """
            latex_code += subfigure_code
        
        latex_code += "\\medskip\n"  # Adding a new line with \medskip at the end of each folder's content
        latex_code += "\\end{figure}\n"  # Close the figure environment for the folder

latex_code += "%====================================% \n"
print(latex_code)
```
